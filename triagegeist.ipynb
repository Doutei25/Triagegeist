```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, f1_score, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import lightgbm as lgb

pd.set_option('display.max_columns', 60)
sns.set_style('whitegrid')
RANDOM_STATE = 42
print("✅ All imports successful")
```

```python
train      = pd.read_csv('train.csv')
test       = pd.read_csv('test.csv')
complaints = pd.read_csv('chief_complaints.csv')
history    = pd.read_csv('patient_history.csv')
submission = pd.read_csv('sample_submission.csv')

print(f"Train:       {train.shape}")
print(f"Test:        {test.shape}")
print(f"Complaints:  {complaints.shape}")
print(f"History:     {history.shape}")
print(f"Submission:  {submission.shape}")

print(f"\nTrain columns: {list(train.columns)}")
print(f"\nComplaints columns: {list(complaints.columns)}")
print(f"\nHistory columns: {list(history.columns)}")
```

```python
# From complaints, only bring in chief_complaint_raw (system already in train)
complaints_merge = complaints[['patient_id', 'chief_complaint_raw']]

# From history, bring all hx_ columns
history_merge = history  # all 26 columns

# Merge into train
train = train.merge(complaints_merge, on='patient_id', how='left')
train = train.merge(history_merge,    on='patient_id', how='left')

# Merge into test
test  = test.merge(complaints_merge, on='patient_id', how='left')
test  = test.merge(history_merge,    on='patient_id', how='left')

print(f"Train shape: {train.shape}")
print(f"Test shape:  {test.shape}")

# Confirm no duplicate columns
dup_cols = [c for c in train.columns if train.columns.tolist().count(c) > 1]
print(f"\nDuplicate columns: {dup_cols}")

# Confirm chief_complaint_raw merged correctly
print(f"\nSample chief complaints:")
print(train[['patient_id','chief_complaint_raw','chief_complaint_system']].head(3).to_string())
```

```python
# Fix pain_score -1 to NaN first
train['pain_score'] = train['pain_score'].replace(-1, np.nan)
test['pain_score']  = test['pain_score'].replace(-1, np.nan)

# Create missingness flags BEFORE filling
missing_cols = ['systolic_bp', 'diastolic_bp', 'respiratory_rate',
                'temperature_c', 'spo2', 'pain_score']

for col in missing_cols:
    train[f'{col}_missing'] = train[col].isnull().astype(int)
    test[f'{col}_missing']  = test[col].isnull().astype(int)

# Verify flags captured the missingness
print("✅ Missingness flags created BEFORE filling")
print("\nMissing counts in train:")
flag_cols = [f'{c}_missing' for c in missing_cols]
for col, flag in zip(missing_cols, flag_cols):
    null_count = train[col].isnull().sum()
    flag_count = train[flag].sum()
    match = "✅" if null_count == flag_count else "❌"
    print(f"  {col:<25} nulls={null_count:>5}  flag={flag_count:>5}  {match}")
```

```python
# Columns to fill with median
vital_fill_cols = ['systolic_bp', 'diastolic_bp', 'mean_arterial_pressure',
                   'pulse_pressure', 'shock_index', 'respiratory_rate',
                   'temperature_c', 'spo2', 'pain_score',
                   'weight_kg', 'height_cm', 'bmi']

# Compute medians from train only (never leak test info)
medians = train[vital_fill_cols].median()

# Apply to both
train[vital_fill_cols] = train[vital_fill_cols].fillna(medians)
test[vital_fill_cols]  = test[vital_fill_cols].fillna(medians)

# Verify
remaining_train = train[vital_fill_cols].isnull().sum().sum()
remaining_test  = test[vital_fill_cols].isnull().sum().sum()

print(f"✅ Missing values filled using train medians")
print(f"\nRemaining nulls in train: {remaining_train}")
print(f"Remaining nulls in test:  {remaining_test}")
print(f"\nMedians used for filling:")
print(medians.round(2).to_string())
```

```python
# Clinical risk flags
train['hypoxia_flag']          = (train['spo2'] < 94).astype(int)
test['hypoxia_flag']           = (test['spo2'] < 94).astype(int)

train['tachycardia_flag']      = (train['heart_rate'] > 100).astype(int)
test['tachycardia_flag']       = (test['heart_rate'] > 100).astype(int)

train['hypotension_flag']      = (train['systolic_bp'] < 90).astype(int)
test['hypotension_flag']       = (test['systolic_bp'] < 90).astype(int)

train['fever_flag']            = (train['temperature_c'] > 38.3).astype(int)
test['fever_flag']             = (test['temperature_c'] > 38.3).astype(int)

train['altered_consciousness'] = (train['gcs_total'] < 15).astype(int)
test['altered_consciousness']  = (test['gcs_total'] < 15).astype(int)

train['tachypnea_flag']        = (train['respiratory_rate'] > 20).astype(int)
test['tachypnea_flag']         = (test['respiratory_rate'] > 20).astype(int)

train['high_pain']             = (train['pain_score'] >= 8).astype(int)
test['high_pain']              = (test['pain_score'] >= 8).astype(int)

print("✅ Part 1 done")
```

```python
# Arrival and shift flags (raw strings before encoding)
train['high_risk_arrival'] = train['arrival_mode'].isin(['ambulance','helicopter']).astype(int)
test['high_risk_arrival']  = test['arrival_mode'].isin(['ambulance','helicopter']).astype(int)

train['is_night']          = (train['shift'] == 'night').astype(int)
test['is_night']           = (test['shift'] == 'night').astype(int)

# Comorbidity burden
hx_cols = [c for c in train.columns if c.startswith('hx_')]
train['total_comorbidities'] = train[hx_cols].sum(axis=1)
test['total_comorbidities']  = test[hx_cols].sum(axis=1)

# Shock index
train['shock_index_calc'] = train['heart_rate'] / train['systolic_bp'].replace(0, np.nan)
test['shock_index_calc']  = test['heart_rate']  / test['systolic_bp'].replace(0, np.nan)

# Age flags
train['elderly_flag']   = (train['age'] >= 65).astype(int)
test['elderly_flag']    = (test['age'] >= 65).astype(int)

train['pediatric_flag'] = (train['age'] < 18).astype(int)
test['pediatric_flag']  = (test['age'] < 18).astype(int)

print("✅ Part 2 done")

# Summary
flag_cols = ['hypoxia_flag','tachycardia_flag','hypotension_flag','fever_flag',
             'altered_consciousness','tachypnea_flag','high_pain',
             'high_risk_arrival','is_night','elderly_flag','pediatric_flag']

summary = pd.DataFrame({
    'count': train[flag_cols].sum(),
    'pct':   (train[flag_cols].mean() * 100).round(1)
})
print(f"\nClinical flags summary:")
print(summary.to_string())

print(f"\nAvg comorbidities by ESI level:")
print(train.groupby('triage_acuity')['total_comorbidities'].mean().round(2).to_string())
```

```python
le = LabelEncoder()

cat_cols = ['arrival_mode', 'arrival_day', 'arrival_season', 'shift',
            'sex', 'language', 'insurance_type', 'transport_origin',
            'mental_status_triage', 'chief_complaint_system',
            'age_group', 'site_id', 'triage_nurse_id', 'pain_location']

encoded = []
for col in cat_cols:
    if col in train.columns and col in test.columns:
        combined = pd.concat([train[col], test[col]], axis=0).astype(str)
        le.fit(combined)
        train[col] = le.transform(train[col].astype(str))
        test[col]  = le.transform(test[col].astype(str))
        encoded.append(col)

print(f"✅ Encoded {len(encoded)} columns")

# Verify no unwanted object columns remain
obj_cols = [c for c in train.columns
            if train[c].dtype == 'object'
            and c not in ['patient_id', 'chief_complaint_raw', 'disposition']]
print(f"\nRemaining object columns (should be empty): {obj_cols}")

print(f"\nTrain shape: {train.shape}")
print(f"Test shape:  {test.shape}")
```

```python
# Fill any missing complaints
train['chief_complaint_raw'] = train['chief_complaint_raw'].fillna('unknown')
test['chief_complaint_raw']  = test['chief_complaint_raw'].fillna('unknown')

# TF-IDF vectorizer — fit on train only
tfidf = TfidfVectorizer(
    max_features=300,
    ngram_range=(1, 2),
    min_df=5,
    strip_accents='unicode',
    lowercase=True
)

tfidf_train = tfidf.fit_transform(train['chief_complaint_raw'])
tfidf_test  = tfidf.transform(test['chief_complaint_raw'])

print(f"TF-IDF matrix shape — Train: {tfidf_train.shape}, Test: {tfidf_test.shape}")

# Reduce dimensions with SVD (LSA)
svd = TruncatedSVD(n_components=30, random_state=RANDOM_STATE)
tfidf_train_svd = svd.fit_transform(tfidf_train)
tfidf_test_svd  = svd.transform(tfidf_test)

print(f"After SVD — Train: {tfidf_train_svd.shape}, Test: {tfidf_test_svd.shape}")
print(f"Variance explained by 30 components: {svd.explained_variance_ratio_.sum():.2%}")

# Add SVD components as new columns
svd_cols = [f'complaint_svd_{i}' for i in range(30)]
train_svd_df = pd.DataFrame(tfidf_train_svd, columns=svd_cols, index=train.index)
test_svd_df  = pd.DataFrame(tfidf_test_svd,  columns=svd_cols, index=test.index)

train = pd.concat([train, train_svd_df], axis=1)
test  = pd.concat([test,  test_svd_df],  axis=1)

print(f"\n✅ NLP features added")
print(f"Train shape: {train.shape}")
print(f"Test shape:  {test.shape}")

# Top words per ESI level
print("\nTop TF-IDF terms associated with ESI 1 (most urgent):")
esi1_idx = train[train['triage_acuity'] == 1].index
esi1_tfidf = tfidf_train[esi1_idx].mean(axis=0).A1
top_terms = np.argsort(esi1_tfidf)[::-1][:10]
vocab = tfidf.get_feature_names_out()
print([vocab[i] for i in top_terms])
```

```python
# Columns to drop (not useful for prediction)
drop_cols = [
    'patient_id',           # ID only
    'chief_complaint_raw',  # replaced by SVD features
    'disposition',          # train-only outcome, would cause leakage
    'ed_los_hours',         # train-only outcome, would cause leakage
    'triage_acuity'         # target
]

# Build feature list
feature_cols = [c for c in train.columns if c not in drop_cols]

# Final matrices
X_train = train[feature_cols]
y_train = train['triage_acuity']
X_test  = test[[c for c in feature_cols if c in test.columns]]

print(f"✅ Final feature matrix ready")
print(f"X_train: {X_train.shape}")
print(f"X_test:  {X_test.shape}")
print(f"y_train distribution:\n{y_train.value_counts().sort_index().to_string()}")

# Confirm no nulls
print(f"\nNulls in X_train: {X_train.isnull().sum().sum()}")
print(f"Nulls in X_test:  {X_test.isnull().sum().sum()}")

# Confirm feature alignment
missing_in_test = [c for c in feature_cols if c not in test.columns]
print(f"\nFeatures in train but not test: {missing_in_test}")
```

```python
# Class weights to handle imbalance
class_counts = y_train.value_counts().sort_index()
total = len(y_train)
class_weights = {cls: total / (len(class_counts) * count)
                 for cls, count in class_counts.items()}

print("Class weights:")
for cls, w in class_weights.items():
    print(f"  ESI {cls}: {w:.3f}")

# LightGBM parameters
params = {
    'objective':         'multiclass',
    'num_class':         5,
    'metric':            'multi_logloss',
    'learning_rate':     0.05,
    'num_leaves':        127,
    'max_depth':         -1,
    'min_child_samples': 20,
    'feature_fraction':  0.8,
    'bagging_fraction':  0.8,
    'bagging_freq':      5,
    'lambda_l1':         0.1,
    'lambda_l2':         0.1,
    'verbose':           -1,
    'random_state':      RANDOM_STATE
}

# Arrays to store predictions
oof_preds  = np.zeros((len(X_train), 5))
test_preds = np.zeros((len(X_test), 5))
fold_scores = []

print("✅ Setup ready")
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

print("Training with 5-Fold Cross Validation...\n")

for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, y_train)):
    X_tr  = X_train.iloc[train_idx]
    X_val = X_train.iloc[val_idx]
    y_tr  = y_train.iloc[train_idx]
    y_val = y_train.iloc[val_idx]

    # LightGBM needs 0-indexed labels
    y_tr_lgb  = y_tr - 1
    y_val_lgb = y_val - 1

    # Sample weights for class imbalance
    sample_weights = y_tr.map(class_weights)

    dtrain = lgb.Dataset(X_tr,  label=y_tr_lgb, weight=sample_weights)
    dval   = lgb.Dataset(X_val, label=y_val_lgb)

    model = lgb.train(
        params,
        dtrain,
        num_boost_round=1000,
        valid_sets=[dval],
        callbacks=[
            lgb.early_stopping(stopping_rounds=50, verbose=False),
            lgb.log_evaluation(period=200)
        ]
    )

    val_pred = model.predict(X_val)
    oof_preds[val_idx] = val_pred
    test_preds += model.predict(X_test) / 5

    val_classes = np.argmax(val_pred, axis=1) + 1
    score = f1_score(y_val, val_classes, average='macro')
    fold_scores.append(score)
    print(f"Fold {fold+1} — Macro F1: {score:.4f} | Best iter: {model.best_iteration}")

print("\n✅ Training complete!")
print(f"Mean Macro F1: {np.mean(fold_scores):.4f}")
print(f"Std:           {np.std(fold_scores):.4f}")
```

```python
# Convert OOF predictions to class labels
oof_classes = np.argmax(oof_preds, axis=1) + 1

# Overall metrics
print("=" * 50)
print("OUT-OF-FOLD EVALUATION (full training set)")
print("=" * 50)
print(f"Macro F1:    {f1_score(y_train, oof_classes, average='macro'):.4f}")
print(f"Accuracy:    {(y_train == oof_classes).mean():.4f}")
print()
print("Per-class Report:")
print(classification_report(y_train, oof_classes,
      target_names=['ESI 1','ESI 2','ESI 3','ESI 4','ESI 5']))

# Confusion matrix
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

cm = confusion_matrix(y_train, oof_classes)
cm_pct = cm.astype(float) / cm.sum(axis=1, keepdims=True) * 100

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['ESI 1','ESI 2','ESI 3','ESI 4','ESI 5'],
            yticklabels=['ESI 1','ESI 2','ESI 3','ESI 4','ESI 5'])
axes[0].set_title('Confusion Matrix (Counts)', fontsize=13)
axes[0].set_ylabel('True Label')
axes[0].set_xlabel('Predicted Label')

sns.heatmap(cm_pct, annot=True, fmt='.1f', cmap='Blues', ax=axes[1],
            xticklabels=['ESI 1','ESI 2','ESI 3','ESI 4','ESI 5'],
            yticklabels=['ESI 1','ESI 2','ESI 3','ESI 4','ESI 5'])
axes[1].set_title('Confusion Matrix (% per True Class)', fontsize=13)
axes[1].set_ylabel('True Label')
axes[1].set_xlabel('Predicted Label')

plt.tight_layout()
plt.show()
```

```python
# Feature importance from last fold model
importance_df = pd.DataFrame({
    'feature':    X_train.columns,
    'importance': model.feature_importance(importance_type='gain')
}).sort_values('importance', ascending=False)

# Top 30 features
top30 = importance_df.head(30)

fig, ax = plt.subplots(figsize=(10, 12))
bars = ax.barh(top30['feature'][::-1], top30['importance'][::-1], color='steelblue')
ax.set_title('Top 30 Most Important Features (LightGBM Gain)', fontsize=13)
ax.set_xlabel('Feature Importance (Gain)')

# Color clinical flags differently
colors = []
for feat in top30['feature'][::-1]:
    if any(x in feat for x in ['news2','gcs','spo2','heart','bp','respiratory',
                                 'shock','temp','svd','complaint']):
        colors.append('steelblue')
    elif feat.startswith('hx_'):
        colors.append('coral')
    else:
        colors.append('mediumseagreen')

for bar, color in zip(bars, colors):
    bar.set_color(color)

from matplotlib.patches import Patch
legend = [Patch(color='steelblue', label='Vitals / NLP'),
          Patch(color='coral',      label='Comorbidities'),
          Patch(color='mediumseagreen', label='Demographics / Context')]
ax.legend(handles=legend, loc='lower right')

plt.tight_layout()
plt.show()

print("Top 10 features:")
print(top30[['feature','importance']].head(10).to_string())
```

```python
# Convert test predictions to class labels (add 1 back since we subtracted earlier)
final_preds = np.argmax(test_preds, axis=1) + 1

# Build submission dataframe
sub = pd.DataFrame({
    'patient_id':    test['patient_id'],
    'triage_acuity': final_preds
})

# Verify format matches sample submission
print("✅ Submission preview:")
print(sub.head(10).to_string())

print(f"\nSubmission shape: {sub.shape}")
print(f"\nPrediction distribution:")
print(sub['triage_acuity'].value_counts().sort_index().to_string())

print(f"\nSample submission distribution (for comparison):")
print(submission['triage_acuity'].value_counts().sort_index().to_string())

# Save
sub.to_csv('submission.csv', index=False)
print("\n✅ submission.csv saved!")
```
